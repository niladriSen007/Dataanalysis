```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

# Line Plot 


```python
var_1 = [1,2,3,4,5,6]
var_2 = [2,3,1,4,5,6]
plt.plot(var_1,var_2)
plt.show()
```

```python
dataframe = {"var_1":var_1,"var_2":var_2}
sns.lineplot(x="var_1",y="var_2",data=dataframe)
```

```python

sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=dataset,hue="sex",style="sex",palette="Dark2",markers=["o",">"],)
plt.show()
```
