```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Line Plot 


```python
var_1 = [1,2,3,4,5,6]
var_2 = [2,3,1,4,5,6]
plt.plot(var_1,var_2)
plt.show()
```

```python
dataframe = {"var_1":var_1,"var_2":var_2}
sns.lineplot(x="var_1",y="var_2",data=dataframe)
```

```python
dataset = sns.load_dataset("penguins").head(200)
dataset
```

```python

sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=dataset,hue="sex",style="sex",palette="Dark2",markers=["o",">"],)
plt.show()
```

# Bar Plot

```python
dataset = sns.load_dataset("penguins")
dataset
```

```python
sns.barplot(x="island",y="bill_length_mm",data=dataset,hue="sex",palette="flare",hue_order=["Female","Male"],errcolor="g")
```

# Histogram

```python
dataset = sns.load_dataset("penguins")
dataset
```

```python
sns.displot(dataset["bill_length_mm"],color="r",kde=True)
```

# Scatter

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=dataset,hue="sex",palette="spring",style="sex",size="sex",sizes=(70,40),markers=True)
plt.plot()
```

# Heatmap

```python
var = np.linspace(1,10,20).reshape(4,5)
sns.heatmap(data=var)
plt.show()
```
